```python
while True:
  try:
   valor_compra = float(input('Insira o valor da compra: '))
   print(f'\n ---> O valor da compra é {valor_compra:.2f} R$')
   print('-------------------------------------------------')

   opcao = int(input('Deseja continuar? \n\n 1 - Para continuar \n 2 - Para encerrar\n\nInsira sua resposta: '))
   print('-------------------------------------------------\n')

   if opcao == 1:
    continue


   elif opcao == 2:
    print('Programa encerrado...')
    print('=============================================')
    break

   else:
            print('[ERRO] Valor digitado é inválido. Digite 1 ou 2.')
            print('-------------------------------------------------\n')

  except ValueError:
        print('-------------------------------------------------\n')
        print('[ERRO] Digite apenas números válidos.')
        print('-------------------------------------------------\n')


```
